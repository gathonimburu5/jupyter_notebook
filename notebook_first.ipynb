# Genarate Fibonacci

```python
def gen_fibonacci(x):
    fib_sequence = [0, 1]
    for _ in range(2, x):
        next_fib = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_fib)
    return fib_sequence
print(gen_fibonacci(10))
```

# Genarate Student Grading

```python
def generate_student_mark(std_name, std_marks_dict, max_subject_mark = 100):
    total_marks = sum(std_marks_dict.values())
    subject_number = len(std_marks_dict)

    max_total_marks = subject_number * max_subject_mark

    if subject_number == 0:
        return { "Name": std_name, "Total Mark": 0, "Percentage": 0, "Grade": "N/A" }

    percentage = (total_marks/max_total_marks) * 100

    if percentage >= 85:
        grade = "A"
    elif percentage >= 75:
        grade = "B+"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 65:
        grade = "B-"
    elif percentage >= 50:
        grade = "C+"
    elif percentage >= 45:
        grade = "C"
    elif percentage >= 40:
        grade = "C-"
    elif percentage >= 35:
        grade = "D+"
    elif percentage >= 30:
        grade = "D"
    elif percentage >= 25:
        grade = "D-"
    else:
        grade = "E"

    return { "Name": std_name, "Total Marks": total_marks, "Percentage": round(percentage, 2), "Grade": grade }

#student1
student1_name = "N8N MBURU"
student1_marks = { "Math": 85, "Science": 92, "Business": 80, "English": 78 }

#student2
student2_name = "JANE GATHONI"
student2_marks = { "Math": 75, "Science": 82, "Business": 70, "English": 68 }

#student3
student3_name = "MAKEBA KAMAU"
student3_marks = { "Math": 45, "Science": 72, "Business": 50, "English": 88 }

#results
result1 = generate_student_mark(student1_name, student1_marks)
result2 = generate_student_mark(student2_name, student2_marks)
result3 = generate_student_mark(student3_name, student3_marks)

print("___________________________________________________________________________________________________________________________")
print("--- Result for N8N MBURU ---")
print(result1)

print("___________________________________________________________________________________________________________________________")
print("--- Result for JANE GATHONI ---")
print(result2)

print("___________________________________________________________________________________________________________________________")
print("--- Result for MAKEBA KAMAU ---")
print(result3)
print("___________________________________________________________________________________________________________________________")
```
