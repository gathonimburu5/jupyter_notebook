```python
numbers = [i for i in range(5, 201, 5)]
print(numbers)
```

```python
import numpy as np
grades = np.array(numbers)
print(grades)
```

```python
study_minutes = [i for i in range(5, 201) if i % 5 == 0]
student_data = np.array([study_minutes, grades])
print(student_data)
```

```python
student_data[0][5]
```

```python
avg_minutes = student_data[0].mean()
avg_grade = student_data[1].mean()
print(f"Average Study Minutes: {avg_minutes}, Average Grade: {avg_grade}")
```

```python
import pandas as pd
data_student = pd.DataFrame({
    "student_names": [f"Student {i}" for i in range(1, 41)],
    "Study Minutes": student_data[0],
    "Grades": student_data[1]
})
print(data_student)
```
