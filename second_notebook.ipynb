```python
study_minutes = [i for i in range(5, 201) if i % 5 == 0]
student_data = np.array([study_minutes, grades])
print(student_data)
```

```python
import pandas as pd
grade_results = pd.read_csv("student_data.csv")
passed_student = pd.Series(grade_results.student_marks >= 50)
grade_results = pd.concat([grade_results, passed_student.rename("Passed")], axis=1)
print(grade_results)
```

```python
grade_results.groupby(grade_results.Passed).student_name.count()
```

```python
grade_results.groupby(grade_results.Passed).student_marks.mean()
```

```python
grade_results.groupby(grade_results['Passed'])[['hours_study', 'student_marks']].mean()
```

```python
grade_results = grade_results.sort_values(['student_marks'])
print(grade_results)
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(20, 4))
ax[0].bar(x=grade_results.student_name, height=grade_results.student_marks)
ax[0].set_xlabel("Student Names")
ax[0].set_ylabel("Student Marks")
ax[0].set_title("Student Marks Bar Chart")
ax[0].set_xticklabels(grade_results.student_name, rotation=90)

passed_count = grade_results.Passed.value_counts()
ax[1].pie(passed_count, labels=passed_count.index, autopct='%1.1f%%', startangle=90)
ax[1].set_title("Passed vs Failed Students")

fig.show()
```

```python
grade_results.plot.bar(x='student_name', y='student_marks', title='Student Marks Bar Chart', figsize=(12,6), rot=90, color='teal')
```

```python
histo_result = grade_results.student_marks

min_val = grade_results.student_marks.min()
max_val = grade_results.student_marks.max()
mean_val = grade_results.student_marks.mean()
median_val = grade_results.student_marks.median()
mode_val = grade_results.student_marks.mode()[0]
print(f"Min: {min_val}, Max: {max_val}, Mean: {mean_val}, Median: {median_val}, Mode: {mode_val}")

fig = plt.figure(figsize=(10,6))
plt.hist(histo_result)

plt.axvline(min_val, color='blue', linestyle='dashed', linewidth=1, label='Min')
plt.axvline(max_val, color='red', linestyle='dashed', linewidth=1, label='Max')
plt.axvline(mean_val, color='green', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median_val, color='orange', linestyle='dashed', linewidth=1, label='Median')
plt.axvline(mode_val, color='purple', linestyle='dashed', linewidth=1, label='Mode')
plt.legend()

plt.xlabel("Student Marks")
plt.ylabel("Frequency")
plt.title("Student Marks Histogram")
plt.show()
```

```python
grade_result = grade_results.student_marks
fig = plt.figure(figsize=(10,6))
plt.boxplot(grade_result, vert=False)
plt.xlabel("Student Marks")
plt.title("Student Marks Boxplot")
plt.show()
```
